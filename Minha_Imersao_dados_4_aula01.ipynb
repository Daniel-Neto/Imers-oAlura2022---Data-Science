```python
# Descobrir quais bairros não tem nome de rua. Nessa linha de código sabe-se que são 3434 instâncias sem rua.

sem_rua = dados["Rua"].isna()
```

```python
# Com isso dá pra saber quantos bairros estão com instâncias sem rua
len(dados[sem_rua]["Bairro"].unique())
```

```python
# Aqui mostra os bairros que estão sem rua e a recorrência de instâncias sem rua por bairro
dados[sem_rua]["Bairro"].value_counts()
```
