```python
# Deixar o gráfico anterior legível

import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(10, 8))
ax = sns.histplot(data = dados_vendas, x = "Valor_anuncio_float", kde = True, color="lightgreen")
ax.set_title("Histograma Valor Imóvel")
# Isso aqui serve para limitar a escala de valores que vai aparecer. Portanto vão aparecer valores de 0 a 10 milhões
plt.xlim((-50, 10000000))
# Essa linha de código faz com que os ticklabel format ignorem o 1e7, deixando o número inteiro
plt.ticklabel_format(style = 'plain')
plt.xlabel("Valor Imóvel")
plt.ylabel("Quantidade")
plt.show()
```
