```python
# Organizando a base do IBGE
ibge_sp = pd.read_csv(
    'https://gist.githubusercontent.com/tgcsantos/85f8c7b0a2edbc3e27fcad619b37d886/raw/a4954781e6bca9cb804062a3eea0b3b84679daf4/Basico_SP1.csv',
    encoding='utf8',
    sep=';', thousands='.', decimal=','
) # thousands='.' troca a vírgula por ponto na milhar decimal=',' coloca , nos decimais
ibge_sp.dropna(how='all', axis=1, inplace=True)
ibge_sp.head()
```

```python
# Esse código faz o tratamento das colunas, fazendo com que eliminem-se espaços vazios
ibge_sp.columns = ibge_sp.columns.str.strip()
# Renomeia a coluna para o nome correto
ibge_sp.rename(columns={'Cod_Grandes Regi�es': "Cod_Grandes Regiões"}, inplace=True)
```

```python
ibge_sp["Nome_da_UF"]
```

```python
# Aqui é possível dar replace em todos os valores errados para colocar ã em vez do caractere estranho. Isso é feito para todas as colunas que possuem string
for column in ibge_sp.columns:
  if type(ibge_sp[column][0]) is str:
    ibge_sp[column] = ibge_sp[column].str.replace("�", "ã")
```

Agora vamos ler a base de dados de endereço, que possui latitudes e longitudes. Essa informação será útil para obter o código do setor de geolocalização.

```python
enderecos = pd.read_csv('/content/drive/MyDrive/Imersão_Dados_Alura/enderecos.csv')
```

```python
enderecos.head()
```

```python
# Pega só os endereços de São Paulo
enderecos_sp = enderecos[enderecos["sigla_uf"] == "SP"]
enderecos_sp
# Outra maneira de pegar somente os endereços de São Paulo: enderecos_sp = enderecos.query("sigla_uf == 'SP'")
```

```python
# Criando uma variável Rua, que é a junção do tipo logradouro com o nome logradouro
enderecos_sp["rua"] = enderecos_sp["tipo_logr"] + " " + enderecos_sp["logr_nome"]
enderecos_sp["rua"] = enderecos_sp["rua"].str.lower().str.strip() # Aqui estou passando o nome para menúsculo e retirando espaços em branco do início e do fim dos nomes
enderecos_sp.head()
```

```python
dados_vendas.head(2)
```

```python
dados_vendas["apenas_rua"] = dados_vendas["Rua"].str.extract(r'(^[\w ]+)')
dados_vendas["apenas_rua"] = dados_vendas["apenas_rua"].str.lower().str.strip()
dados_vendas.head(2)
```

```python
# Esse código está fazendo um left join da tabela dados_vendas com enderecos_sp. Isso vai retornar linhas repetidas, por isso, faz-se uma remoção de duplicatas
# quando o cep > 0. Desse modo, só ficará uma linha por rua.
dados_geo = pd.merge(left = dados_vendas, right = enderecos_sp[["rua", "cep", "latitude", "longitude"]], how = "left", left_on = "apenas_rua", right_on = "rua").drop_duplicates(subset=dados_vendas.columns).query("cep > 0")
```

```python
# Agora, vamos ler o arquivo de dados que possuem o setor censo, através do geopandas
setor_censo = gpd.read_file('/content/drive/MyDrive/Imersão_Dados_Alura/35SEE250GC_SIR.shp')
setor_censo.head(2)
```

```python
# Essa base tem várias informações, inclusive o Polygon na coluna geometry. A partir disso é possível verificar se um ponto
# está presente numa região censitária
# Vamos pegar cada latitude e longitude da base de dados e retornar o seu setor censitário

setor_censo_sp = setor_censo[setor_censo['NM_MUNICIP'] == "SÃO PAULO"] # Pega somente os setores censitários de São Paulo
setor_censo_sp[setor_censo_sp.contains(Point(longitude,latitude))] # Retorna somente a linha que possui aquela longitude e latitude

# Nesse caso o interesse é retornar o CD_GEOCODI... isso que será utilizado para a base do IBGE
```

Agora vamos pegar cada ponto e verificar se este contém no setor censo, retornando o seu GEOCODE e criando uma coluna nova na base dados_geo.

```python
dados_geo["Point"] = ""
for i in dados_geo.index:
  dados_geo["Point"][i] = Point(dados_geo["longitude"][i], dados_geo["latitude"][i])

# Faz a função map e retorna os valores de CD_GEOCODI
dados_geo["setor_censo"] = dados_geo["Point"].map(lambda x: setor_censo_sp.loc[setor_censo_sp.contains(x), 'CD_GEOCODI'].values).str[0]
dados_geo
```

Agora vamos obter as linhas dessa base cujo setor_censo não seja NaN

```python
dados_geo = dados_geo[dados_geo["setor_censo"].notnull()]
dados_geo
```

Após isso, podemos fazer o merge final. Da base dados_geo com a base do ibge, gerando uma base só.

```python
ibge_sp.info()
```

```python
# é necessário transformar o setor_censo para int64, para realizar o merge
dados_geo["setor_censo"] = dados_geo["setor_censo"].astype("int64")
```

```python
dados_vendas_censo = pd.merge(left = dados_geo, right = ibge_sp, how = "left", left_on = "setor_censo", right_on = "Cod_setor")
dados_vendas_censo
```
